```python
import pandas as pd
path="/content/drive/MyDrive/dataset/wael1.csv"
data=pd.read_csv(path)
data.head()
df=pd.DataFrame(data)
summary = df.describe()
print(summary)
df=df.dropna(axis=1)
print(df)
df["order_date"] = pd.to_datetime(df["order_date"], format='mixed')
df.drop_duplicates(inplace = True)
print(df.duplicated())
first_row = df.iloc[0]
last_row = df.iloc[-1]
print("First Row:")
print(first_row)
print("\nLast Row:")
print(last_row)
df.head(1000)
```

```python
import numpy as np
np.array(df)
age_mean=df[('age')].astype(int).mean()
print("the average age of buying a product ",age_mean)
quantity_mean=df[('quantity')].astype(int).mean()
print("The average of quantity bought",quantity_mean)
price_mean=df[('price')].astype(int).mean()
print("The Average Price ",price_mean)
mostly_used=df[('payment_method')].mode()
print("the most payment method used ",mostly_used)
cat_name=df[('category_name')].mode()
print("the most category used ",cat_name)
sum_quantity=df[('quantity')].sum()
print("the sum of quantity ",sum_quantity)
sum_price=df[('price')].sum()
print("the sum of price ",sum_price)
products=df[('product_name')].mode()
print("the most peoduct used ",products)
filtered_df = df[df['payment_method'] == 'Credit Card']
citym=filtered_df[('city')].mode()
print("the most payment method used ",citym)
```

```python
import pandas as pd

def clean_csv(filepath, output_path=None):
    df = pd.read_csv('/content/drive/MyDrive/dataset/wael1.csv')
    df.dropna(axis=1, how='all', inplace=True)
    df.dropna(how='all', inplace=True)
    df.columns = df.columns.str.strip()
    df.drop_duplicates(inplace=True)
    df['order_date'] = pd.to_datetime(df['order_date'], format='mixed')
    df.reset_index(drop=True, inplace=True)
    if 'Content' in df.columns:
        df = df.dropna(subset=['Content'])
    df.reset_index(drop=True, inplace=True)
    if output_path:
        df.to_csv(output_path, index=False)
        print(f"Cleaned CSV saved to {output_path}")
    return df
cleaned_df = clean_csv("wael1.csv", "wael1_cleaned.csv")

```
